# Matplotlib example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig = plt.figure()
fig, ax = plt.subplots(figsize=(5, 5))

# 3. Plot data
plt.plot(x, y)

# 4. Customize plot
ax.set(title="SampleSimplePlot", xlabel="x-axis", ylabel="y-axis")

# Save and show
fig.savefig("Images/sample_plot.png")
```

## Two Options for subplots

```python
#Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

#Plot to each differernt axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
#plot to each differnt index
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

### Pyplot vs Matplotlib OO Method
* When plotting quickly use pyplot methods
* When plotting something advanced use OO method

```python
# heart_disease.query("age > 50") can also be used
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
# Pyplot method
over_50.plot(kind='scatter', x='age', y='chol', c='target')
```

```python
# OO Method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax);
#ax.set_xlim([0,100])
```

```python
# Pure OO(Object Oriented) method from scratch
fig, ax = plt.subplots(figsize=(10,6))

#Plot data
scatter = ax.scatter(x=over_50["age"], y= over_50["chol"], c=over_50["target"]);

#Customize
ax.set(title="Heart Disease and Cholestrol Level", xlabel="Age", ylabel="Cholestrol")

#Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

#Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle="--")
```

```python
#Sublot of chol age , thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10))

#Add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])
#Customize ax0
ax0.set(title="Heart Disease cholestrol and Cholestrol level", xlabel="Age", ylabel="Cholestrol")
#Add a legend to x0
ax0.legend(*scatter.legend_elements(), title="target")
#Add meanline
ax0.axhline(y=over_50["chol"].mean(), linestyle="--")

scatter = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"])
ax1.set(title="Heart Disease and Heartrate level", xlabel="age", ylabel="heartrate")
ax1.legend(*scatter.legend_elements(), title="target")
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--")


#Add title to figure

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
#Sublot of chol age , thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10))

#Add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap="winter") #cmap means legend color
#Customize ax0
ax0.set(title="Heart Disease cholestrol and Cholestrol level", xlabel="Age", ylabel="Cholestrol")

ax0.set_xlim([50,80]) #Change x axis limit ie crop graph properly

#Add a legend to x0
ax0.legend(*scatter.legend_elements(), title="target")
#Add meanline
ax0.axhline(y=over_50["chol"].mean(), linestyle="--")

scatter = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"], cmap="magma") #cmap means legend color
ax1.set(title="Heart Disease and Heartrate level", xlabel="age", ylabel="heartrate")
ax1.set_xlim([40,100]) #Change x axis limit ie crop graph properly
ax1.set_ylim([80,200])
ax1.legend(*scatter.legend_elements(), title="target")
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--")


#Add title to figure

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
