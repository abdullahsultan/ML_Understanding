# Matplotlib example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig = plt.figure()
fig, ax = plt.subplots(figsize=(5, 5))

# 3. Plot data
plt.plot(x, y)

# 4. Customize plot
ax.set(title="SampleSimplePlot", xlabel="x-axis", ylabel="y-axis")

# Save and show
fig.savefig("Images/sample_plot.png")
```

## Making figures with NumPy Arrays

We want:
* Line plot
* Scatter plot
* Bar plot
* histogram
* Subplots

```python
import numpy as np
```

```python
#Create some data
x = np.linspace(0, 10, 100, dtype=float)
x
```

```python
#Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use same data to make scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictonary
nut_butter_prices = {
    "Almond Butter": 10,
    "Peanut Butter": 8,
    "Cachew Butter": 12
}
fig, ax = plt.subplots()
ax.set(title="NutButter", ylabel="Price$")
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
```

```python
# Make some data for histograms and plot in it
x=np.random.randn(1000)
fig, ax=plt.subplots()
ax.hist(x);
```

## Two Options for subplots

```python
#Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

#Plot to each differernt axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
#plot to each differnt index
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

## Plotting from pandas dataframe

```python
import pandas as pd
```

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
# Example
ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2023", periods=1000))
ts = ts.cumsum()
ts.plot()
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '', regex=True)
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Sale Data"]= pd.date_range("1/1/2023", periods=len(car_sales))
```

```python
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
```

```python
car_sales
```

```python
# Plot total sales
car_sales.plot(x="Sale Data", y="Total Sales");
```

```python
car_sales["Price"] = car_sales["Price"].astype(int)
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter");
```

```python
car_sales
```

```python
car_sales.plot.bar(x="Odometer (KM)", y="Price");
```

```python
car_sales.groupby("Make").plot.bar(x="Make", y="Price");
```
