# Matplotlib example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

#1 Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#2 Setup plot
fig
ax = plt.subplot(figsize=(5,10))

#3 Plot data
ax.plot(x,y)

#4 Customize plot
ax.set("Simple Plot", xlabel="x-axis", ylabel="y-axis")

# Save and show
fig.savefig("Images/samepl_plot.png")
```
