# Matplotlib example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig = plt.figure()
fig, ax = plt.subplots(figsize=(5, 5))

# 3. Plot data
plt.plot(x, y)

# 4. Customize plot
ax.set(title="SampleSimplePlot", xlabel="x-axis", ylabel="y-axis")

# Save and show
fig.savefig("Images/sample_plot.png")
```

## Making figures with NumPy Arrays

We want:
* Line plot
* Scatter plot
* Bar plot
* histogram
* Subplots

```python
import numpy as np
```

```python
#Create some data
x = np.linspace(0, 10, 100, dtype=float)
x
```

```python
#Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use same data to make scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictonary
nut_butter_prices = {
    "Almond Butter": 10,
    "Peanut Butter": 8,
    "Cachew Butter": 12
}
fig, ax = plt.subplots()
ax.set(title="NutButter", ylabel="Price$")
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
```
